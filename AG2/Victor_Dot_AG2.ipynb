```
import random
import math

N=4000
LISTA_2D = [(random.randrange(1,N*10),random.randrange(1,N*10)) for _ in range(N)]

print(LISTA_2D)
```

```
#Decorador para calcular_tiempo a la función para la cual queremos calcular el tiempo de ejecución(Isabel Vázquez)

from time import time
#Función para calcular el tiempo de ejecución
def calcular_tiempo(f):
   
    def wrapper(*args, **kwargs):        
        inicio = time()       
        resultado = f(*args, **kwargs)       
        tiempo = time() - inicio
        print("Tiempo de ejecución para algoritmo: "+str(tiempo))
        return resultado
    
    return wrapper
```

```
def distancia(A,B):
  if type(A) is int or type(A) is float:
    return abs(B-A)
  else:
    return math.sqrt(sum([   (A[i]-B[i])**2 for i in range(len(A))   ]))
  

distancia ((1,3),(2,5))
```

```
#fuerza bruta


def distancia_fuerza_bruta(L):
  mejor_distancia_conseguida = 10000e10
  
  A,B = (), ()
  
  for i in range(len(L)):
    for j in range(i+1, len(L)):
      if distancia(L[i],L[j]) < mejor_distancia_conseguida:
        A,B=L[i],L[j]
        mejor_distancia_conseguida = distancia(L[i],L[j])
  return(A,B)

distancia_fuerza_bruta(LISTA_2D) 

```

```
def distancia_divide_y_venceras(L):
  #Si la lista es corta lo calculamos por fuerza bruta
  if len(L)<10:
    return distancia_fuerza_bruta(L)
  
  #Divide y venceras
  #pivote =  sum([L[i][0]for i in range(len(L))]) / len(L)
  
  LISTA_IZQ = sorted(L, key=lambda x: x[0])[:len(L)//2]
  LISTA_DER = sorted(L, key=lambda x: x[0])[len(L)//2:]
  
  PUNTOS_LISTA_IZQ = distancia_divide_y_venceras(LISTA_IZQ)
  PUNTOS_LISTA_DER = distancia_divide_y_venceras(LISTA_DER)
  
  return distancia_fuerza_bruta(LISTA_IZQ + LISTA_DER)

@calcular_tiempo
def LANZA(L):
  return distancia_divide_y_venceras(L)

SOL = LANZA(LISTA_2D[:100])

print(SOL)
```

```
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

#Paseo por el rio
def Precios(TARIFAS):
  N = len(TARIFAS[0])
  
  PRECIOS = [ [9999]*N for i in [9999]*N ]
  RUTAS    = [ [""]*N for i in [9999]*N ]
  

  
  for i in range(N-1):
    for j in range(i+1,N):
      MIN = TARIFAS[i][j]
      RUTAS[i][j] = i
      
      for k in range(i,j):
        if PRECIOS[i][k]+ TARIFAS[k][j] < MIN:
          MIN = min( MIN , PRECIOS[i][k]+ TARIFAS[k][j] )
          RUTAS[i][j] = k
      PRECIOS[i][j] = MIN
      
  return PRECIOS, RUTAS

  
PRECIOS, RUTAS = Precios(TARIFAS)

print(PRECIOS)

print()

print(RUTAS)

def calcular_ruta(RUTAS, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return desde 
  else:
    return str(calcular_ruta(RUTAS, desde, RUTAS[desde][hasta])) +  ',' + str(RUTAS[desde][hasta]) 

print("\nLa ruta es:")  
calcular_ruta(RUTAS, 0,6)
```
